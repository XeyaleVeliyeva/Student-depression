```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/Student Depression Dataset.csv")
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df = df.dropna()
```

```python
df.isna().sum()
```

```python
df["City"].unique()
```

Delete unavailable city names


```python
cities = [
    'Saanvi', 'M.Tech', 'Bhavna', 'Less Delhi', 'City', '3',
    'Less than 5 Kalyan', 'Mira', 'Harsha', 'Vaanya', 'Gaurav',
    'Harsh', 'Reyansh', 'Kibara', 'Rashi', 'ME', 'M.Com',
    'Nalyan', 'Mihir', 'Nalini', 'Nandini', 'Khaziabad'
]

df = df.drop(df[df["City"].isin(cities)].index).reset_index(drop=True)
```

Correlation heatmap


```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True),annot=True,cmap="twilight",lw=0.7)
```

```python
sns.countplot(x="Gender",hue="Depression",data = df,palette="rocket")
plt.title("Depression by Gender")
```

```python
depressed=df[df["Depression"]==1]
sns.histplot(data=depressed,x="Age",bins=20,color="#2459be",kde=True)
plt.title("Depressed people's age histogram")
```

```python
depressed
```

```python
gender = depressed["Gender"].value_counts()
plt.pie(gender,labels = gender.index,autopct="%1.1f%%",colors=["skyblue","hotpink"]);
plt.title("Depression percentage by Gender");
```

```python
sns.boxplot(data = df,x="Age")
plt.title("Outliers in age column");
```

Depressed people across cities

```python
depressed = df[df["Depression"]==1]
depression_count = depressed.groupby("City")["Depression"].value_counts()
depression_count
```

```python
depression_count_cities = depression_count.sort_values(ascending=False).head(10)
depression_count_cities
```

```python
a = depression_count_cities.reset_index()
a
```

```python
# depression_count_cities_1 = depression_count_cities.reset_index(drop=True)
sns.barplot(data=a,x="City",y="count",palette = "mako");
plt.xticks(rotation=90);
```

```python
sns.countplot(data=df,x="Dietary Habits",hue="Depression", palette="magma")

plt.title("Impact of Dietary Habits on Depression")
plt.xlabel("Health'")
plt.ylabel("Count")
plt.legend(title='Group')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.countplot(data=df,x="Degree",hue="Depression", palette="mako")

plt.title("Impact of Degrees on Depression")
plt.xlabel("Degree'")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend(title='Group')
plt.show()
```

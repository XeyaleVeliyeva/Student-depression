```python
total_counts = df['Have you ever had suicidal thoughts ?'].value_counts()
total_counts
```

```python
depressed_counts = df[df['Depression'] == 1]['Have you ever had suicidal thoughts ?'].value_counts()
depressed_counts
```

```python
depression_percentage = (depressed_counts / total_counts) * 100
depression_percentage
```

```python
import matplotlib.pyplot as plt
depression_percentage.plot(kind='bar', color=['red', 'green'], alpha=0.7)

plt.title('Depression Percentage by Suicidal Thoughts')
plt.xlabel('Have you ever had suicidal thoughts?')
plt.ylabel('Percentage of Depressed People')
plt.xticks(rotation=0)
plt.show()
```

```python
depressed = df[df['Depression'] == 1]
non_depressed = df[df['Depression'] == 0]

depressed_counts = depressed['Have you ever had suicidal thoughts ?'].value_counts()
non_depressed_counts = non_depressed['Have you ever had suicidal thoughts ?'].value_counts()

non_depressed_counts
```

```python
import matplotlib.pyplot as plt
import pandas as pd

suicidal_thoughts_df = pd.DataFrame({
    'Depressed': depressed_counts,
    'Non-Depressed': non_depressed_counts
})

suicidal_thoughts_df.plot(kind='bar', figsize=(8, 6), color=['red', 'green'])

plt.title("Depressed vs Non-Depressed")
plt.xlabel("Response to 'Have you ever had suicidal thoughts?'")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.legend(title='Group')
plt.show()
```

```python
suicidal_thoughts_gender_depressed = depressed.groupby('Gender')['Have you ever had suicidal thoughts ?'].value_counts()
# suicidal_thoughts_gender_depressed
suicidal_thoughts_gender_non_depressed = non_depressed.groupby('Gender')['Have you ever had suicidal thoughts ?'].value_counts()
# suicidal_thoughts_gender_non_depressed

suicidal_thoughts_gender = pd.DataFrame({
    'Depressed': suicidal_thoughts_gender_depressed,
    'Non-Depressed': suicidal_thoughts_gender_non_depressed
})
colors = sns.color_palette('crest',len(suicidal_thoughts_gender))
suicidal_thoughts_gender.plot(kind='bar', figsize=(8, 6), color=colors)
plt.title("Depressed vs Non-Depressed")
plt.xlabel("Response to 'Have you ever had suicidal thoughts?'")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.legend(title='Group')
plt.show()
```

```python
city = df.groupby('City')['Financial Stress'].value_counts().reset_index(name='Count')
city
```

```python
pivot_data = city.pivot(index='City', columns='Financial Stress', values='Count').fillna(0)

pivot_data.plot(
    kind='bar',
    stacked=True,
    figsize=(15, 8),
    colormap='viridis'
)

plt.title('Financial Stress Counts by City (Stacked)')
plt.xlabel('City')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.legend(title='Financial Stress')
plt.show()
```
